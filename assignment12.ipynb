Model `mpg` as a function of `displacement` and `cylinders`. 

- Transform the `displacement` variable using a `spline` or `polynomial` transform. (Hint: use `step_ns` or `step_poly`).
- Combine `step_mutate(cyl_cat = factor(cylinders))` followed by `step_dummy(cyl_cat)` and `step_rm(cylinders)` to convert `cylinders` to categorical data for the model.

**Create and visualize your model below...**

````r
# Define the recipe
rec <- recipe(mpg ~ displacement + cylinders, data = Auto) |>
  step_ns(displacement, deg_free = 3) |>  # Apply spline transformation on displacement
  step_mutate(cyl_cat = factor(cylinders)) |>  # Convert cylinders to categorical
  step_dummy(cyl_cat) |>  # Create dummy variables for categorical cylinders
  step_rm(cylinders)      # Remove the original cylinders column

# Define the model
mod <- linear_reg() |>
  set_engine("lm")

# Fit the model
mod_fit <- workflow() |>
  add_recipe(rec) |>
  add_model(mod) |>
  fit(data = Auto)

# Make predictions using the original dataset
predictions <- predict(mod_fit, new_data = Auto) |>
  bind_cols(Auto)

# Visualize the results
ggplot(predictions, aes(x = displacement, y = mpg)) +
  geom_point(size = 1, alpha = 0.35, color = "gray") +  # Scatter plot for actual data
  geom_line(aes(y = .pred, color = factor(cylinders)), linewidth = 1) +  # Predicted values
  scale_color_brewer(palette = "Set1", name = "Cylinders") +  # Color by cylinders
  labs(
    title = "Modeling MPG as a Function of Displacement and Cylinders",
    x = "Displacement (cubic inches)",
    y = "Miles Per Gallon (MPG)"
  ) +
  theme_minimal()

````
